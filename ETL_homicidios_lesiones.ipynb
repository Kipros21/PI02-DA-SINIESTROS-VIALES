```python
#Definimos algunas funciones para que nos facilita la descripcion de las principales caracteristicas del DataFrame
def caracteristicas_df(df):
    """
    Describe de forma general la base de datos .

    Esta función simplemente muestra el tamaño, información general y
    la cantidad de datos nulos.

    Parametros
    ----------
    df (pandas.DataFrame): El DataFrame que se va a analizar.

    Returns:
    ----------
        - 'df.shape': Numero de filas y columnas
        - 'df.info': Muestra información general del DataFrame

    """
    print('*'*10 + '|'*10 + 'FORMA DE BASE DE DATOS' + '|'*10 + '*'*10, end = '\n'*2)
    print(f'Tiene {df.shape[0]} filas y {df.shape[1]} columnas o variables')
    print(end = '\n'*2)

    print('*'*10 + '|'*10 + 'INFORMACION GENERAL DE LA BASE DE DATOS' + '|'*10 + '*'*10, end = '\n'*2)
    print(df.info(), end = '\n'*2)

def valores_nulos_df(df):
    """
    Revisa presencia de valores nulos en un DataFrame.
    Esta función toma un DataFrame como entrada y devuelve un resumen que incluye información sobre
    el porcentaje de valores no nulos y nulos, así como la ncantidad de valores nulos por columna.

    Parametros:
    ----------
    df (pandas.DataFrame): El DataFrame que se va a analizar.

    Returns:
    ----------
        pandas.DataFrame: Un DataFrame que contiene el resumen de cada columna, incluyendo:
        - 'nombre': Nombre de cada columna.
        - 'no_nulos_%': Porcentaje de valores no nulos en cada columna.
        - 'nulos_%': Porcentaje de valores nulos en cada columna.
        - 'nulos': Cantidad de valores nulos en cada columna.

    """
    mi_df = {"nombre": [], "tipo_datos": [], "nulos_%": [], "nulos": []}

    for columna in df.columns:
        porcentaje_no_nulos = (df[columna].count() / len(df)) * 100
        mi_df["nombre"].append(columna)
        mi_df["tipo_datos"].append(df[columna].apply(type).unique())
        mi_df["nulos_%"].append(round(100-porcentaje_no_nulos, 2))
        mi_df["nulos"].append(df[columna].isnull().sum())

    df_nulos = pd.DataFrame(mi_df)

    return df_nulos
```

### 2.3. Transformación de Datos

#### 2.3.1. Funciones de Transformación

```python
def rango_etario(edad):
    """
    Asigna un rango etario basado en la edad proporcionada.

    Parameters:
    -----------
        - edad (int): Edad de la persona.

    Returns:
    -----------
        - str or np.nan: Etiqueta del rango etario o NaN si no corresponde a ninguno.
    """
    # Definimos los rangos etarios y sus respectivas etiquetas
    rangos_etarios = {
        (0, 2): "BEBE",
        (3, 12): "NIÑO",
        (13, 17): "ADOLESCENTE",
        (18, 29): "ADULTO JOVEN",
        (30, 39): "ADULTO MEDIANA EDAD",
        (40, 49): "ADULTO MADURO",
        (50, 59): "ADULTO MAYOR",
        (60, float('inf')): "TERCERA EDAD"
    }

    # Iteramos sobre los rangos etarios
    for rango, etiqueta in rangos_etarios.items():
        # Verificamos si la edad se encuentra en el rango actual
        if rango[0] <= edad <= rango[1]:
            return etiqueta

    # Si la edad no coincide con ningún rango, retornamos NaN
    return np.nan

def estado_gravedad(victima):
    """
    Genera una carátula basada en el tipo de victima.

    Parameters:
    -----------
        - gravedad (str): Nivel de gravedad ('LEVE', 'GRAVE' o 'FATAL').

    Returns:
    -----------
        - str: Carátula correspondiente.
    """
    # Paso 1: Verificar el nivel de gravedad y generar la carátula correspondiente
    if victima in ['CICLISTA', 'PEATON', 'MOTO', 'MONOPATIN', 'UTILITARIO']:
        return 'LESIONES GRAVES'
    elif victima in ['TAXI', 'AUTO', 'MOVIL', 'MIXTO', 'CAMIONETA']:
        return 'LESIONES MODERADAS'
    elif victima in ['TRANSPORTE PUBLICO', 'SD', 'OTRO', 'CAMION']:
        return 'LESIONES LEVES'
    else:
        # Paso 2: Manejo de casos no contemplados (otros niveles de gravedad)
        return 'NO DEFINIDO'

def sentencia(gravedad):
    """
    Genera una sentencia basada en el nivel de gravedad.

    Parameters:
    -----------
        - gravedad (str): Nivel de gravedad ('LESIONES LEVES', 'LESIONES MODERADAS' o 'LESIONES GRAVES').

    Returns:
    -----------
        - str: Carátula correspondiente.

    Steps:
    -----------
        1. Se verifica el nivel de gravedad proporcionado.
        2. Si el nivel de gravedad es 'LESIONES LEVES', se genera la carátula 'FALTA'.
        3. Si el nivel de gravedad es 'LESIONES MODERADAS', se genera la carátula 'DELITO LEVE'.
        4. Si el nivel de gravedad es 'LESIONES GRAVES', se genera la carátula 'DELITO GRAVE'.
        5. En caso de niveles de gravedad no contemplados, se devuelve 'NO DEFINIDO'.
    """
    # Paso 1: Verificamos el nivel de gravedad y generar la carátula correspondiente
    if gravedad == 'LESIONES LEVES':
        return 'FALTA'
    elif gravedad == 'LESIONES MODERADAS':
        return 'DELITO LEVE'
    elif gravedad == 'LESIONES GRAVES':
        return 'DELITO GRAVE'
    else:
        # Paso 2: Manejamos de casos no contemplados (otros niveles de gravedad)
        return 'NO DEFINIDO'
    
def verifica_duplicados(df, columna):
    '''
    Verifica y muestra filas duplicadas en un DataFrame basado en una columna específica.

    Esta función toma como entrada un DataFrame y el nombre de una columna específica.
    Luego, identifica las filas duplicadas basadas en el contenido de la columna especificada,
    las filtra y las ordena para una comparación más sencilla.

    Parameters:
    ----------
        df (pandas.DataFrame): El DataFrame en el que se buscarán filas duplicadas.
        columna (str): El nombre de la columna basada en la cual se verificarán las duplicaciones.

    Returns:
    ----------
        pandas.DataFrame or str: Un DataFrame que contiene las filas duplicadas filtradas y ordenadas,
        listas para su inspección y comparación, o el mensaje "No hay duplicados" si no se encuentran duplicados.
    '''
    # Se filtran las filas duplicadas
    duplicated_rows = df[df.duplicated(subset=columna, keep=False)]
    if duplicated_rows.empty:
        return "No hay duplicados"

    # se ordenan las filas duplicadas para comparar entre sí
    duplicated_rows_sorted = duplicated_rows.sort_values(by=columna)
    return duplicated_rows_sorted

from datetime import datetime

def convertir_a_time(x):
    '''
    Convierte un valor a un objeto de tiempo (time) de Python si es posible.

    Parameters:
    -----------
        x (str, datetime, or any): El valor que se desea convertir a un objeto de tiempo (time).

    Returns:
    -----------
        datetime.time or None: Un objeto de tiempo (time) de Python si la conversión es exitosa,
        o None si no es posible realizar la conversión.

    Raises:
    -----------
        ValueError: Si la conversión a tiempo falla debido a un formato incorrecto.

    Steps:
    -----------
        1. Se verifica si el valor es una cadena (str).
        2. Se intenta convertir la cadena a un objeto de tiempo (time) utilizando `datetime.strptime`.
        3. Si la conversión es exitosa, se devuelve el objeto de tiempo.
        4. Si la conversión no es posible debido a un `ValueError`, se maneja devolviendo `None`.
        5. Se verifica si el valor es un objeto `datetime`.
        6. Si es un objeto `datetime`, se obtiene la parte de tiempo del objeto.
        7. Si ninguna de las conversiones es posible, devuelve el valor original.
    '''
    if isinstance(x, str):
        try:
            return datetime.strptime(x, "%H:%M:%S").time()
        except ValueError:
            return None
    elif isinstance(x, datetime):
        return x.time()
    
    return x

def imputa_valor_frecuente(df, columna):
    '''
    Imputa los valores faltantes en una columna de un DataFrame con el valor más frecuente.

    Esta función reemplaza los valores "SD" con NaN en la columna especificada,
    luego calcula el valor más frecuente en esa columna y utiliza ese valor
    para imputar los valores faltantes (NaN).

    Parameters:
    -----------
        df (pandas.DataFrame): El DataFrame que contiene la columna a ser imputada.
        columna (str): El nombre de la columna en la que se realizará la imputación.

    Returns:
    -----------
        None

    Raises:
    -----------
        ValueError: Si la columna especificada no existe en el DataFrame.

    Steps:
    -----------
        1. Se reemplaza "SD" con NaN en la columna.
        2. Se calcula el valor más frecuente en la columna.
        3. Se imputan los valores NaN con el valor más frecuente.
    '''
    # Verificar si la columna especificada existe en el DataFrame
    if columna not in df.columns:
        raise ValueError(f'La columna "{columna}" no existe en el DataFrame.')

    # Reemplazar "SD" con NaN en la columna
    df[columna] = df[columna].replace(['SD', 'sd'], pd.NA)

    # Calcular el valor más frecuente en la columna
    valor_mas_frecuente = df[columna].mode().iloc[0]
    print(f'El valor más frecuente en la columna "{columna}" es: {valor_mas_frecuente}')

    # Imputar los valores NaN con el valor más frecuente
    df[columna].fillna(valor_mas_frecuente, inplace=True)

def imputa_edad_media_segun_sexo(df):
    '''
    Imputa valores faltantes en la columna 'Edad' utilizando la edad promedio según el género.

    Esta función reemplaza los valores "SD" con NaN en la columna 'Edad', calcula la edad promedio
    para cada grupo de género (Femenino y Masculino), imprime los promedios calculados y
    luego llena los valores faltantes en la columna 'Edad' utilizando el promedio correspondiente
    al género al que pertenece cada fila en el DataFrame.

    Parameters:
    -----------
        df (pandas.DataFrame): El DataFrame que contiene la columna 'Edad' a ser imputada.

    Returns:
    -----------
        None

    Raises:
    -----------
        ValueError: Si la columna 'Edad' no existe en el DataFrame.

    Steps:
    -----------
        1. Se reemplaza "SD" con NaN en la columna 'Edad'.
        2. Se calcula el promedio de edad para cada grupo de género.
        3. Se imprime la edad promedio calculada para Femenino y Masculino.
        4. Se llenan los valores NaN en la columna 'Edad' utilizando el promedio correspondiente al género.
        
    '''
    # Verificar si la columna 'Edad' existe en el DataFrame
    if 'Edad' not in df.columns:
        raise ValueError('La columna "Edad" no existe en el DataFrame.')

    # Reemplazar "SD" o "sd" con NaN en la columna 'Edad'
    df['Edad'] = df['Edad'].replace(['SD', 'sd'], pd.NA)

    # Calcular el promedio de edad para cada grupo de género
    promedio_por_genero = df.groupby('Sexo')['Edad'].mean()
    print(f'La edad promedio de Femenino es {round(promedio_por_genero["FEMENINO"])} y de Masculino es {round(promedio_por_genero["MASCULINO"])}')

    # Llenar los valores NaN en la columna 'Edad' utilizando el promedio correspondiente al género
    df['Edad'] = df.apply(lambda row: promedio_por_genero[row['Sexo']] if pd.isna(row['Edad']) else row['Edad'], axis=1)

    # Convertir la columna 'Edad' a tipo entero
    df['Edad'] = df['Edad'].astype(int)

def transformar_direccion(texto):
    """
    Reordena y normaliza las columnas referentes a las direcciones.

    Esta función toma un texto como entrada y realiza las siguientes transformaciones:
    - Convierte el texto a mayúsculas.
    - Reordena el texto en el formato deseado, eliminando la coma y normalizando abreviaciones.
    - Reemplaza abreviaciones específicas con sus formas completas.

    Parameters:
    -----------
        texto (str): El texto de la columna a transformar.

    Returns:
    -----------
        str or np.nan: El texto transformado o NaN si el input no es una cadena.

    Steps:
    -----------
        1. El texto de la columna se convierte a mayúsculas para asegurar coherencia en el formato.
        2. Si el texto contiene una coma, se divide en partes y se reordena según el formato deseado. Por ejemplo, "Calle A, Ciudad" se convierte en "Ciudad Calle A".
        3. Se reemplazan abreviaciones específicas, como 'GRAL.' por 'GENERAL', 'TTE.' por 'TENIENTE', y 'CNEL.' por 'CORONEL'.
        4. Si el texto contiene 'DR.', se divide en partes y se reordena. Lo mismo se aplica para 'AV.'.
        5. Se eliminan espacios en blanco adicionales alrededor del texto.
        6. Si el input no es una cadena, se retorna NaN.       
    
    """
    if isinstance(texto, str):
        # Convertir a mayúsculas
        texto = texto.upper()

        # Reordenar el texto
        if "," in texto:
            partes = [parte.strip() for parte in texto.split(",")]
            if len(partes) == 2:
                texto = f"{partes[1]} {partes[0]} "
            elif len(partes) == 3:
                texto = f"{partes[2]} {partes[1]} {partes[0]} "

        # Tratar abreviaciones específicas
        texto = texto.replace('GRAL.', 'GENERAL').replace('TTE.', 'TENIENTE').replace('CNEL.', 'CORONEL')

        # Tratar 'DR.' y 'AV.'
        if "DR." in texto:
            partes_dr = [parte.strip() for parte in texto.split("DR.")]
            if len(partes_dr) == 2:
                texto = f"DR. {partes_dr[0]} {partes_dr[1]}"
        
        if "AV." in texto:
            partes_av = [parte.strip() for parte in texto.split("AV.")]
            if len(partes_av) == 2:
                texto = f"AV. {partes_av[0]} {partes_av[1]}"

        # Eliminar espacios en blanco
        texto = texto.strip()

        return texto
    else:
        # Manejar casos no cadena
        return float(np.nan)

def reemplazar_participacion(value):
    '''
    Reemplaza valores en las columnas designadas para los participantes del siniestro.

    Esta función busca reemplazar valores NaN en las columnas designadas para los participantes
    del siniestro, como ['MOTO', 'AUTO', 'TRANSPORTE_PUBLICO', 'CAMION', 'CICLISTA'].

    Parameters:
    -----------
        value (str): El valor actual en la columna a ser reemplazado.

    Returns:
    -----------
        float or str: El valor reemplazado. Si el valor es 'x', se reemplaza por 1.0 (float redondeado). 
                     Si el valor es '0', se reemplaza por 0.0 (float redondeado). En otros casos, 
                     se devuelve el valor original.
    '''
    if value == 'x':
        # Si el valor es 'x', se reemplaza por 1 (float redondeado)
        return 1
    elif value == '0':
        # Si el valor es '0', se reemplaza por 0 (float redondeado)
        return 0
    else:
        # Si el valor no es 'x' ni '0', se devuelve 0
        return 0
```
